```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('../cleaned.log', sep=';')

share = pd.pivot_table(df, index='timestamp', columns='node', values='renewable_share')
score = pd.pivot_table(df, index='timestamp', columns='node', values='score')

index = pd.date_range(start = "2020-06-01 00:00", end = "2020-06-01 23:50", freq = "10min")
index = [pd.to_datetime(date, format='%Y-%m-%d %H:%M') for date in index]

share['timestamp'] = index
share = share.set_index('timestamp')

score['timestamp'] = index
score = score.set_index('timestamp')

ax = share.plot(figsize=(12,4),color = ['lightblue', 'lightcoral', 'teal'])
plt.xlabel("Time")
plt.ylabel("Renewable Energy Share in Watts")

score.plot(figsize=(12,4),color = ['lightblue', 'lightcoral', 'teal'])

#ax.axvline("2020-06-01 23:10:00 ", color="green", linestyle="dashed")

plt.xlabel("Time")
plt.ylabel("Score")

plt.legend(loc='lower right',title="node")
plt.show()
```
